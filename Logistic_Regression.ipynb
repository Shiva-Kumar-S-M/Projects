```python
#Binomial Logistic regression

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=23)

reg=LogisticRegression(max_iter=10000,random_state=0)
reg.fit(X_train,y_train)

y_pred=reg.predict(X_test)
acc=accuracy_score(y_test,y_pred)*100

print(f"Logistic regression accuracy score:{acc:.2f}%")
```

```python
#Multinomial Logistic regression
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score,classification_report


digits=load_digits()
X=digits.data
y=digits.target

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=1)

model=LogisticRegression(max_iter=10000,random_state=1)
model.fit(X_train,y_train)

y_pred=model.predict(X_test)

print("Accuracy Score:",accuracy_score(y_test,y_pred)*100)

print("Precision Score:",precision_score(y_test,y_pred,average='macro')*100)

print("F1 Score:",f1_score(y_test,y_pred,average='macro')*100)

print("Recall Score:",recall_score(y_test,y_pred,average='macro')*100)

print("Classification report :",classification_report(y_test,y_pred))

```

```python
#Softmax Functionn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
data=load_iris()
X=data.data
y=data.target

y_encoded=to_categorical(y)

X_train,X_test,y_train,y_test=train_test_split(X,y_encoded,test_size=0.2,random_state=42)

model=Sequential([
    Dense(8,input_shape=(4,),activation='relu'),
    Dense(3,activation='softmax')
])

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

history=model.fit(X_train,y_train,epochs=100,batch_size=8,validation_split=0.2,verbose=0)

loss,accuracy=model.evaluate(X_test,y_test)
print("Test Accuracy:",accuracy)
```
